```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
```

```python
r = requests.get('https://www.audaxindia.org/event-e-2352')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
date_string = re.findall('on (.+)', soup.find(id="banner").div.find_all(attrs = {"class":"row"})[1].h1.text)[0]
```

```python
date_object = datetime.datetime.strptime(date_string, '%d-%b-%Y')
```

```python
data = []
```

```python
for record_soup in soup.find(id="fellowriders").table.tbody.find_all('tr'):
    record = {}
    record['name'] = record_soup.find_all('td')[1].text.strip()
    record['air_number'] = record_soup.find_all('td')[2].text.strip()
    time = record_soup.find_all('td')[5].text.strip()
    if(re.search('.+h.+m', time)):
        hrs = re.findall('(.+)h', time)
        mins = re.findall(' (.+)m', time)
        record['time'] = int(hrs[0]) * 60 + int(mins[0])
    else:
        record['time'] = time
    record['club'] = record_soup.find_all('td')[6].text.strip()
    data.append(record)
print(data[0])
```

```python
DISTANCE = 200
```

```python
for record_soup in soup.find(id="fellowriders").table.tbody.find_all('tr'):
    record = {}
    record['name'] = record_soup.find_all('td')[1].text.strip()
    record['air_number'] = record_soup.find_all('td')[2].text.strip()
    time = record_soup.find_all('td')[5].text.strip()
    if(re.search('.+h.+m', time)):
        hrs = re.findall('(.+)h', time)
        mins = re.findall(' (.+)m', time)
        record['time'] = int(hrs[0]) * 60 + int(mins[0])
        record['speed'] = DISTANCE/(record['time']/60)
        record['time_hrs_mins'] = time
    else:
        record['time'] = time
        record['speed'] = 'NA'
        record['time_hrs_mins'] = time
    record['club'] = record_soup.find_all('td')[6].text.strip()
    data.append(record)
print (len(data))
```

```python
group_by_club = data.groupby('club').count()['air_number']
group_by_club = group_by_club.sort_values(ascending=False)
sns.barplot(group_by_club.values, group_by_club.index)
```

```python
group_by_club = data.groupby('time').count()['air_number']
group_by_club = group_by_club.sort_values(ascending=False)
sns.barplot(group_by_club.values, group_by_club.index)
```
